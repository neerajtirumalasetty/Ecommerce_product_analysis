```python
import pandas as pd
```

```python
products = pd.read_csv("../data/products.csv")
reviews = pd.read_csv("../data/reviews.csv")
```

```python
products.head()
```

```python
products[['price_value', 'rating_count', 'rating_stars', 'recent_purchases']].head(10)
```

```python
products['rating_count'] = products['rating_count'].str.replace(',','')
products['rating_count'] = products['rating_count'].str.extract('(\d+)')
products['rating_count'] = products['rating_count'].astype(float)
```

```python
products['rating_stars'] = products['rating_stars'].str.extract('(\d+\.\d+)')
products['rating_stars'] = products['rating_stars'].astype(float)
```

```python
def clean_purchases(x):
    if pd.isna(x):
        return 0
    x = x.replace('+ bought', '').strip()
    if 'K' in x:
        return float(x.replace('K', '')) * 1000
    return float(x)

products['recent_purchases'] = products['recent_purchases'].apply(clean_purchases)
```

```python
products[['price_value', 'rating_count', 'rating_stars', 'recent_purchases']].head()
```

```python
products[['price_value', 'rating_count', 'rating_stars', 'recent_purchases']].info()
```

```python
# feature eng
products['demand_score'] = products['rating_count'] + products['recent_purchases']
```

```python
products['price_bucket'] = pd.cut(
    products['price_value'],
    bins = [0, 20, 40, 100],
    labels = ['Low', 'Medium', 'High']
)
```

```python
import numpy as np

products['log_demand'] = np.log1p(products['demand_score'])
```

```python
products[['price_value', 'demand_score', 'price_bucket', 'log_demand']].head()
```

```python
# insights
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Q1. does price effect demand ?
plt.figure(figsize=(8,5))
sns.scatterplot(data=products, x='price_value', y='demand_score')
plt.title("Price vs Demand")
plt.xlabel("Price")
plt.ylabel("Demand Score")
plt.show()
```

```python
# Q2. which price range performs better ?
products.groupby('price_bucket')['demand_score'].mean()
```

```python
# Q3. do better ratings drive demand ?

plt.figure(figsize=(8,5))
sns.scatterplot(data=products, x='rating_stars', y='demand_score')
plt.title("Rating vs Demand")
plt.xlabel("Rating")
plt.ylabel("Demand Score")
plt.show()
```

```python
# Q4. Demand distribution across price buckets
sns.boxplot(data=products, x='price_buckets', y='demand_score')
plt.title("Demand Ditribution by Price Bucket")
plt.show()
```

```python
# Q5. Top products

products.sort_values(by='demand_score', ascending=False)[
    ['title','price_value', 'demand_score']
].head(10)
```

```python
products['main_category'] = products['breadcrumbs'].str.split('>').str[0]
products['main_category'] = products['main_category'].str.strip()
```

```python
category_demand = products.groupby('main_category')['demand_score'].mean().sort_values(ascending=False)
category_demand
```

```python
plt.figure(figsize=(10,6))
category_demand.head(10).sort_values().plot(kind='barh')

plt.title("Top Categories by Demand")
plt.xlabel("Average Demand")
plt.ylabel("Category")

plt.tight_layout()
plt.show()
```

```python
products['brand_name'] = products['brand_name'].fillna('Unknown')
```

```python
brand_demand = products.groupby('brand_name')['demand_score'].mean().sort_values(ascending=False)

brand_demand.head(10)
```

```python
plt.figure(figsize=(10,6))
brand_demand.head(10).sort_values().plot(kind='barh')

plt.title("Top Brands by Demand")
plt.xlabel("Average Demand")

plt.tight_layout()
plt.show()
```
